<p>This project was made for the completion of <b>IBM Data Science Professional Certificate: Applied Data Science Capstone.</b>

## Business Problem

<p>Tourism is an important sector for Philippine economy. In 2019, the travel and tourism industry contributed 12.7% to the country's GDP. Being one of the fastest growing provinces in the Philippines, Iloilo has also marked itself as one of the most loved province by tourists. The province itself has already taken its spot as one of the tourist hub where great choices of attractions and activities can be found; from restaurants, museums, hotels, beaches, shopping districts, night life and a whole lot more to explore. 
<br /><br />Being a resident of Passi City, I've always wondered if:
    <br/>- Our tourism is contributing in our city's revenue
    <br/>- If it is known in the tourists globally
    <br/>- What are the top tourist spots that can be found. 
<br/><br/>So for this project, I decided to look into the standing of Passi City's tourism compared to its neighboring cities and municipalities in the whole Iloilo province.
</p>

## Who Will Benefit In This Project

### This project can be a benefit for: 

- Passi City government and tourism sectors: information will be given on how many venue categories are known to tourists from the data gathered in Four Square API.

- Start-ups: insights on the top venue categories known to Passi City itself and its competing cities/municipalities based on revenues

## Data to be Used

### For this project, I will be using these data:

- List of cities and municipalities in Iloilo Province with their zipcode: gathered from "https://sites.google.com/site/departmentofphilippines/philippine-zip-codes/provincial-zip-codes/iloilo-zip-codes" by webscraping


- Latitude and Longitude of city/municipality: gathered from geocoder


- Revenues of each city/municipality: gathered from wikipedia (last updated on 2016) 


- List of venues for each city/municipality: gathered from Foursquare API
